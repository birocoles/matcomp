```python
import matplotlib.pyplot as plt
import numpy as np
from tqdm.notebook import tqdm as tq # produce the progress bar
```

```python
N = np.array([50, 100, 500, 1000, 3000, 5000, 10000])
```

```python
list_time_add_reduce = []
list_time_sum = []
for Ni in tq(N):
    vector = np.ones(Ni)
    # numpy.add.reduce
    time = %timeit -o np.add.reduce(vector)
    list_time_add_reduce.append(time.average)
    # numpy.sum
    time = %timeit -o np.sum(vector)
    list_time_sum.append(time.average)
```

```python
plt.figure(figsize=(10,7))
plt.plot(N, np.asarray(list_time_add_reduce)*1e6, 'bo-', label = 'add.reduce')
plt.plot(N, np.asarray(list_time_sum)*1e6, 'ro-', label = 'sum')
plt.legend(loc = 'best', fontsize = 14)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.xlabel('N', fontsize = 16)
plt.ylabel('Time ($\mu$s)', fontsize = 16)
plt.yscale('log')
plt.grid()
plt.show()
```
